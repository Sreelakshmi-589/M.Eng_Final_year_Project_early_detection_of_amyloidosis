```python
# Importing Libraries

import os
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import roc_auc_score,roc_curve
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Defining paths

DATA_PATH = "C:/Users/Sreelakshmi/00_Final_Project_MENG/Approach 2/TabTransformer/split_data/"
ASSETS_DIR = "C:/Users/Sreelakshmi/00_Final_Project_MENG/Approach 2/TabTransformer/"
MODEL_PATH = os.path.join(ASSETS_DIR, "best_full_model_fixed_tabtransformer_dataset2.keras")
```

```python
# Loading Test Data

print("\n--- Loading Test Data ---")
test_df = pd.read_csv(os.path.join(DATA_PATH, "split_data_test.csv"))
y_test = test_df['amyloidosis'].values
print(f"Test data loaded: {test_df.shape[0]} samples")
```

```python
# Preparing tokenizer and pad sequence

print("\n--- Preparing Tokenizer & Sequences ---")

# Load all codes from saved CSVs to fit the tokenizer
train_val_test_df = pd.concat([pd.read_csv(os.path.join(DATA_PATH, f)) for f in ["split_data_train.csv","split_data_validation.csv","split_data_test.csv"]])
all_codes_list = train_val_test_df['X_codes'].apply(lambda x: x.strip('[]').replace("'", "").split(','))
all_codes_list = all_codes_list.apply(lambda x: [c.strip() for c in x]).tolist()
all_codes_list_flat = [' '.join(codes) for codes in all_codes_list]

tokenizer = tf.keras.preprocessing.text.Tokenizer(oov_token="<unk>")
tokenizer.fit_on_texts(all_codes_list_flat)
vocab_size = len(tokenizer.word_index) + 1
print(f"Tokenizer fitted. Vocabulary size: {vocab_size}")

MAX_SEQUENCE_LENGTH = 200
test_sequences = [' '.join(x) if isinstance(x, list) else '' for x in test_df['X_codes'].apply(eval)]
X_test_pad = pad_sequences(tokenizer.texts_to_sequences(test_sequences), maxlen=MAX_SEQUENCE_LENGTH, padding='post', truncating='post')
```

```python
# Loading trained model
from tensorflow.keras.models import load_model

# It is important to make sure the TransformerBlock class is defined
class TransformerBlock(tf.keras.layers.Layer):
    def __init__(self, embed_dim, num_heads, ff_dim, rate=0.1, **kwargs):
        super(TransformerBlock, self).__init__(**kwargs)
        self.embed_dim, self.num_heads, self.ff_dim, self.rate = embed_dim, num_heads, ff_dim, rate
        self.att = tf.keras.layers.MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)
        self.ffn = tf.keras.models.Sequential([
            tf.keras.layers.Dense(ff_dim, activation="relu"),
            tf.keras.layers.Dense(embed_dim)
        ])
        self.layernorm1 = tf.keras.layers.LayerNormalization(epsilon=1e-6)
        self.layernorm2 = tf.keras.layers.LayerNormalization(epsilon=1e-6)
        self.dropout1 = tf.keras.layers.Dropout(rate)
        self.dropout2 = tf.keras.layers.Dropout(rate)

    def call(self, inputs, training=None):
        attn_output = self.att(inputs, inputs)
        attn_output = self.dropout1(attn_output, training=training)
        out1 = self.layernorm1(inputs + attn_output)
        ffn_output = self.ffn(out1)
        ffn_output = self.dropout2(ffn_output, training=training)
        return self.layernorm2(out1 + ffn_output)

# loading the model and mentioning Keras about the custom layer
model = load_model(MODEL_PATH, custom_objects={'TransformerBlock': TransformerBlock})
print("Model loaded successfully.")
```

```python
# Generate predictions on Test set

print("\n--- Making Predictions on Test Set ---")

# Get predicted probabilities (values between 0 and 1)
pred_probs = model.predict(X_test_pad).flatten()

# Convert probabilities to binary class labels (threshold = 0.5)
pred_labels = (pred_probs > 0.5).astype(int)
```

```python
# Evaluate model performance on Test set

print("\n--- Evaluating Model ---")
accuracy = accuracy_score(y_test, pred_labels)
precision = precision_score(y_test, pred_labels)
recall = recall_score(y_test, pred_labels)
f1 = f1_score(y_test, pred_labels)
auc = roc_auc_score(y_test, pred_probs)
tn, fp, fn, tp = confusion_matrix(y_test, pred_labels).ravel()
specificity = tn / (tn + fp)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall (Sensitivity): {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
print(f"Specificity: {specificity:.4f}")
print(f"AUC-ROC: {auc:.4f}")
```

```python
print("\nClassification Report:\n", classification_report(y_test, pred_labels))
```

```python
# Confusion Matrix
conf_matrix = confusion_matrix(y_test, pred_labels)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=['ATTR-CM', 'No ATTR-CM'],  # predicted labels
            yticklabels=['ATTR-CM', 'No ATTR-CM'])   # actual labels
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title("Confusion Matrix on Test Set - TabTransformer Model(Dataset 2)")
plt.show()
```

```python
# ROC Curve
fpr, tpr, _ = roc_curve(y_test, pred_probs)
plt.figure(figsize=(8,6))
plt.plot(fpr, tpr, label=f'AUC = {auc:.2f}')
plt.plot([0,1], [0,1], linestyle='--', color='gray')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve on Test Set')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

```python
# Saving the prediction report

validation_report_df = pd.DataFrame({
    'patient_id': test_df['patient_id'],
    'true_label': y_test,
    'predicted_probability': pred_probs,
    'predicted_label': pred_labels
})
validation_report_path = os.path.join(ASSETS_DIR, "Final_test_predictions_dataset2.csv")
validation_report_df.to_csv(validation_report_path, index=False)
print(f"\nValidation predictions report saved to: {validation_report_path}")
```
