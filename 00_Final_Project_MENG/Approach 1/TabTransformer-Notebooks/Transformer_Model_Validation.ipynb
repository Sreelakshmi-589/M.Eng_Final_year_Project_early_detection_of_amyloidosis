```python
# Importing Libraries
import pandas as pd
import numpy as np
import torch

import joblib
from sklearn.preprocessing import StandardScaler

from pytorch_tabular import TabularModel

from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import roc_auc_score,roc_curve
import matplotlib.pyplot as plt
import seaborn as sns

import shap
```

```python
# Defining paths
val_data_path = 'C:/Users/Sreelakshmi/00_Final_Project_MENG/Approach 1/Dataset_1/val_data.csv'
model_path = 'C:/Users/Sreelakshmi/00_Final_Project_MENG/Approach 1/TabTransformer/tabtransformer_model'
scaler_path = 'C:/Users/Sreelakshmi/00_Final_Project_MENG/Approach 1/TabTransformer/tabTransformer_scaler.sav'
```

```python
# Load and prepare validation set
val_df = pd.read_csv(val_data_path)         # load validation CSV

target_col = "__target__"
features = val_df.drop(columns=[target_col]).columns.tolist()
```

```python
# Display the DataFrame
print(f"Validation DataFrame shape: {val_df.shape}")
```

```python
# Display first 5 rows
val_df.head(5)
```

```python
# Load saved scaler and scale validation data

scaler = joblib.load(scaler_path)       # load the previously saved scaler
 
val_df_scaled = val_df.copy()
val_df_scaled[features] = scaler.fit_transform(val_df[features])  # apply scaling to validation features
```

```python
# Load saved model directory
loaded_model = TabularModel.load_model(model_path)
```

```python
# Generate predictions on validation set

# Get predicted probabilities (values between 0 and 1)
val_probs_df = loaded_model.predict(val_df_scaled)
val_probs = val_probs_df.values[:, 1]  # Probability for class 1

# Convert probabilities to binary class labels (threshold = 0.5)
val_preds = (val_probs >= 0.5).astype(int)
y_true = val_df_scaled[target_col]
```

```python
# Evaluate model performance on validation set

# Accuracy
acc = accuracy_score(y_true, val_preds)
# Precision
prec = precision_score(y_true, val_preds)
# Recall
rec = recall_score(y_true, val_preds)
# F1-score
f1 = f1_score(y_true, val_preds)
# AUC (ROC)
auc = roc_auc_score(y_true, val_probs)

print("Validation Metrics: TabTransformer Model (Dataset 1)")
print(f"Accuracy:  {acc:.4f}")
print(f"Precision: {prec:.4f}")
print(f"Recall:    {rec:.4f}")
print(f"F1-score:  {f1:.4f}")
print(f"AUC:       {auc:.4f}")

# Classification report
print("\nClassification Report TabTransformer Model (Validation Set - Dataset 1):")
print(classification_report(y_true, val_preds))
```

```python
# Confusion Matrix

cm = confusion_matrix(y_true, val_preds)

plt.figure(figsize=(6,5))
sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=['ATTR-CM', 'No ATTR-CM'],  # predicted labels
    yticklabels=['ATTR-CM', 'No ATTR-CM']   # actual labels
)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - TabTransformer (Validation set - Dataset 1)')
plt.show()
```

```python
fpr, tpr, _ = roc_curve(y_true, val_probs)
plt.figure(figsize=(8,6))
plt.plot(fpr, tpr, label="ROC Curve (AUC = {:.4f})".format(roc_auc_score(y_true, val_probs)))
plt.plot([0,1],[0,1],'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - TabTransformer (Validation set - Dataset 1)")
plt.legend(loc="lower right")
plt.show()
```

```python
# SHAP Explainer 

# Sample 100 rows from the scaled validation dataset for SHAP analysis
# Sampling reduces computation time while still capturing representative patterns
X_sample = val_df_scaled[features].sample(100, random_state=42)

# SHAP requires a function that takes a NumPy array as input and returns model predictions
def model_predict(x_numpy):
    # Convert NumPy array to PyTorch tensor (continuous features)
    x_tensor = torch.tensor(x_numpy, dtype=torch.float32)
    
    # Disable gradient calculation for efficiency
    with torch.no_grad():
        # Pass input through the TabTransformer model
        output = loaded_model.model({"continuous": x_tensor})
    
    # Extract predictions from the model output
    if isinstance(output, dict):
        if "logits" in output:
            pred_tensor = output["logits"]       # raw outputs before softmax
        elif "prediction" in output:
            pred_tensor = output["prediction"]   # predicted class
        else:
            pred_tensor = list(output.values())[0]  # fallback: take first value
    else:
        pred_tensor = output
    
    # Convert tensor to NumPy array for SHAP
    return pred_tensor.detach().numpy()


# Creating a SHAP explainer for TabTransformer predictions
# SHAP will call 'model_predict' on perturbed versions of X_sample
explainer = shap.Explainer(model_predict, X_sample.values)

# Computing SHAP values for X_sample
# max_evals controls the number of feature perturbations to evaluate
shap_values = explainer(X_sample.values, max_evals=2 * len(features) + 1)
```

```python
#Extract SHAP values for Class 1
class_idx = 1
shap_class_1 = shap_values.values[:, :, class_idx]  # Shape: (samples=100, features=1873)
```

```python
# Compute Mean Absolute SHAP per Feature

# Take the absolute SHAP values and average across all samples
# This gives a global measure of feature importance
mean_abs_shap = np.mean(np.abs(shap_class_1), axis=0)  # Shape: (1873,)
```

```python
# Store Results in DataFrame

# Building a DataFrame for feature importance ranking
shap_df = pd.DataFrame({
    'feature': features,
    'mean_abs_shap': mean_abs_shap
})

# Sort features by importance (descending order)
shap_df = shap_df.sort_values('mean_abs_shap', ascending=False)
```

```python
# Visualize Top Features

# Plot top 20 most important features for Class 1
shap_df.head(20).plot.bar(
    x='feature', 
    y='mean_abs_shap', 
    title="Top 20 Feature Importances - ATTR-CM", 
)
plt.show()
```
